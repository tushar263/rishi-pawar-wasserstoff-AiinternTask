```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
from string import punctuation
```

```python

# Specify the directory where nltk_data is located
nltk_data_dir = "C:/Users/abish/nltk_data"  # Adjust this path if needed
nltk.data.path.append(nltk_data_dir)

# Download the necessary resources
nltk.download('punkt', download_dir=nltk_data_dir)
nltk.download('stopwords', download_dir=nltk_data_dir)
nltk.download('wordnet', download_dir=nltk_data_dir)


stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = re.sub(r'\s+', ' ', text)  # Clean spaces, tabs, and newlines
    tokens = word_tokenize(text)       # Tokenize using NLTK
    filtered_tokens = [
        lemmatizer.lemmatize(token.lower())  # Lemmatize and convert to lowercase
        for token in tokens 
        if token.lower() not in stop_words and token not in punctuation
    ]
    return " ".join(filtered_tokens)

preprocess_text(text)
```
